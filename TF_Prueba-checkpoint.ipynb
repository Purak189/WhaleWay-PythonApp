```python
import osmnx as ox
import geopandas as gpd

# Define the place name and fetch the graph
place_name = "Independencia, Lima, Per"
graph = ox.graph_from_place(place_name, network_type="drive")

# Convert the graph to GeoDataFrames
nodes, edges = ox.graph_to_gdfs(graph)

# Fill missing values with default values
# For nodes
nodes.fillna({'ref': 'N/A', 'highway': 'N/A'}, inplace=True)
# Add more attributes if necessary

# For edges
edges.fillna({'ref': 'N/A', 'highway': 'N/A'}, inplace=True)
# Add more attributes if necessary

# Save nodes and edges to CSV files
nodes_path = 'nodes.txt'
edges_path = 'edges.txt'
nodes.to_csv(nodes_path, index=False)
edges.to_csv(edges_path, index=False)

# Print confirmation
print("Nodes and edges have been saved to 'nodes.txt' and 'edges.txt'.")

# Create download links in Jupyter Notebook
from IPython.display import display, HTML
import base64

def create_download_link(file_path, title="Download file"):
    with open(file_path, 'r') as file:
        data = file.read()
    b64 = base64.b64encode(data.encode()).decode()
    href = f'<a href="data:file/txt;base64,{b64}" download="{file_path}">{title}</a>'
    return HTML(href)

# Display download links
display(create_download_link(nodes_path, "Download nodes as CSV"))
```

```python
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt
import geopandas as gpd
import json

# Step 1: Load the road network graph for a specific area in Lima
G = ox.graph_from_place('Independencia, Lima, Peru', network_type='all')

# Step 2: Load additional nodes from the geojson file
with open('filtered_nodes.geojson') as f:
    geojson_data = json.load(f)

# Convert geojson features to GeoDataFrame
gdf = gpd.GeoDataFrame.from_features(geojson_data['features'])

# Step 3: Add these nodes to the graph
for idx, row in gdf.iterrows():
    node_id = row['@id']  # Use the 'id' field directly
    y, x = row.geometry.y, row.geometry.x
    shop_type = row['properties'].get('name', 'N/A') if 'properties' in row else 'N/A'
    G.add_node(node_id, y=y, x=x, shop=shop_type)

# Step 4: Draw the graph and highlight additional nodes
fig, ax = plt.subplots(figsize=(12, 12))
ox.plot_graph(G, ax=ax, node_color='blue', node_size=10, edge_color='gray', show=False, close=False)

# Highlight the additional nodes in red
highlight_nodes = gdf['@id'].tolist()
nc = ['red' if node in highlight_nodes else 'blue' for node in G.nodes()]
ox.plot_graph(G, ax=ax, node_color=nc, node_size=20, edge_color='gray', show=True, close=True)

plt.show()
```
